```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

```python
dfs = []
for i in range(0, 11):  # Assuming your CSV files are named from 1.csv to 10.csv
    filename = f"./run_directory/current_results/{i}.csv"
#     filename = f"./concatenated_results0.csv"
    df = pd.read_csv(filename)
    dfs.append(df)
```

```python
def rtd(num):
    rounded_num = round(num, 3)  # Round to three decimal places
    rounded_str = "{:.3f}".format(rounded_num)  # Format to ensure three decimal places

    # Add trailing zero if needed
    if len(rounded_str.split('.')[1]) < 3:
        rounded_str += '0'

    return rounded_str
```

```python
def get_res(method, metric, train_test, std = False):
    query_result = []
    for res in dfs:
        qr = res[(res['Unnamed: 1'] == method) & (res['metric'] == metric)][train_test]
        query_result.append(float(np.array(qr)[0]))
    if std == 0: return rtd(np.mean(query_result))
    if std:
        query_result_std = []
        for res in dfs:
            qrs = res[(res['Unnamed: 1'] == method) & (res['metric'] == metric+'_std')][train_test]
            query_result_std.append(float(np.array(qrs)[0]))
        return rtd(np.mean(query_result)), rtd(np.mean(query_result_std))
    
def get_res(method, metric, train_test, std = False):
    query_result = []
    for res in dfs:
        qr = res[(res['Unnamed: 1'] == method) & (res['metric'] == metric)][train_test]
        query_result.append(float(np.array(qr)[0]))
    if std == 0: return rtd(np.mean(query_result))
    if std:
        return rtd(np.mean(query_result)), rtd(np.std(query_result))

def get_np(method, metric, train_test, std = False):
    neg, pos = [], []
    for res in dfs:
        qr = res[(res['Unnamed: 1'] == method) & (res['metric'] == metric)][train_test]
        neg.append(float(np.array(qr)[0].split('/')[0]))
        pos.append(float(np.array(qr)[0].split('/')[1]))
    return rtd(np.mean(neg)), rtd(np.mean(pos))

def get_ress(method, metric, train_test):
    a, b = get_res(method, metric, train_test, 1)
    print(a, b)
    return str(a)+"_{"+str(b)+"}"
```

```python
def get_data(method, metric, train_test):
    ress = []
    for res in dfs:
        ress.append(float(np.array(res[(res['Unnamed: 1'] == method) & (res['metric'] == metric)][train_test])))
    return ress

def ttest(bigger_array, smaller_array, name1, name2, alpha = 0.05):
    t_statistic, p_value = stats.ttest_ind(bigger_array, smaller_array)
    p_value /= 2
    if p_value < alpha and t_statistic > 0:
        print("Reject null hypothesis: "+name1+" is significantly greater than "+name2+". p-value:", p_value)
    else:
        print("Fail to reject: "+name1+" is not significantly greater than "+name2+". p-value:", p_value)
```
